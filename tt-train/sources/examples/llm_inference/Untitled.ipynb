```python
import ttml
import numpy as np

seq_len = 800
emb_dim = 384

def test_emb_backward(vocab_size):
    assert emb_dim % 32 == 0
    assert vocab_size % 32 == 0

    hidden = np.random.randn(1, 1, seq_len, emb_dim).astype(np.float32)
    lm_head_weight = np.random.randn(1, 1, vocab_size, emb_dim).astype(np.float32)

    hidden_ttml = ttml.autograd.Tensor.from_numpy(hidden, ttml.Layout.TILE, ttml.autograd.DataType.BFLOAT16)
    lm_head_ttml = ttml.autograd.Tensor.from_numpy(lm_head_weight, ttml.Layout.TILE, 
                                                   ttml.autograd.DataType.BFLOAT16)
    bias = lm_head_weight = np.random.randn(1, 1, 1, vocab_size).astype(np.float32)
    bias = ttml.autograd.Tensor.from_numpy(bias, ttml.Layout.TILE, 
                                                   ttml.autograd.DataType.BFLOAT16)

    logits = ttml.ops.linear.linear_op(hidden_ttml, lm_head_ttml, bias)
    logits.backward(False)
    
# doesn't crash    
#test_emb_backward(80000) 

# TT_THROW: Statically allocated circular buffers on core range [(x=0,y=0) - (x=6,y=7)] 
# grow to 1528096 B which is beyond max L1 size of 1499136 B (assert.hpp:103)
#test_emb_backward(75968) 

# TT_THROW: Statically allocated circular buffers on core range [(x=0,y=0) - (x=6,y=7)] 
# grow to 1544480 B which is beyond max L1 size of 1499136 B (assert.hpp:103)
#test_emb_backward(151936) 

test_emb_backward(75968) 
```
